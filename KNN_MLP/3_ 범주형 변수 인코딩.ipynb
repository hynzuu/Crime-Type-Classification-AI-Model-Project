```python
import pandas as pd
import numpy as np
from IPython.display import display as dp
import warnings
import matplotlib
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

base_train = pd.read_csv('../data/new_train.csv', index_col=0)
validation =pd.read_csv('../data/new_validation.csv', index_col=0)
test =pd.read_csv('../data/new_test.csv', index_col=0)

IQR_delete = pd.read_csv('../data/IQR_delete.csv', index_col=0)
IQR_weight = pd.read_csv('../data/IQR_weight.csv', index_col=0)
Isolation_Forest_delete = pd.read_csv('../data/Isolation_Forest_delete.csv', index_col=0)
Isolation_Forest_weight = pd.read_csv('../data/Isolation_Forest_weight.csv', index_col=0)

dp(base_train,validation,test)
```

```python
# 사용할 인코더
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from category_encoders import BinaryEncoder
```

# Ordinal Encoding
**월, 요일, 시간** 컬럼에 적용할 방식이다. 각 클래스가 순서를 가지기 때문에 이를 선택했다.

그런데 월, 시간은 이미 숫자로 입력되어있고 EDA에서 범주형 변수라고 생각했기 때문에 임의로 내가 category 타입으로 변경해두었을 뿐이다.

따라서 **월**과 **시간**은 number로 바꿔주고 요일에 대해서만 인코딩을 진행할 것

```python
# 인코더는 후자의 방식으로 input값을 설정해야함.
base_train['요일'].shape, base_train[['요일']].shape
```

```python
ordinal_encoder = OrdinalEncoder()
base_train['요일'] = ordinal_encoder.fit_transform(base_train[['요일']])
validation['요일'] = ordinal_encoder.transform(validation[['요일']]) # val/test는 학습된 인코더로 변환만 
test['요일'] = ordinal_encoder.transform(test[['요일']])
```

```python
base_train
```

```python
Isolation_Forest_weight
```

```python
# base의 인코딩 방식을 따르기 위해 각각이 훈련 데이터셋이라도 transform만 함.
IQR_delete['요일'] = ordinal_encoder.transform(IQR_delete[['요일']])
IQR_weight['요일'] = ordinal_encoder.transform(IQR_weight[['요일']])
Isolation_Forest_delete['요일'] = ordinal_encoder.transform(Isolation_Forest_delete[['요일']]) 
Isolation_Forest_weight['요일'] = ordinal_encoder.transform(Isolation_Forest_weight[['요일']])
```

# Binary Encoding
소관경찰서, 소관 지역, 범죄발생지 컬럼에 적용할 방식이다. 고유한 클래스가 굉장히 많으므로 효과적일 것이라 예상된다.  
예를 들어 8개의 범주가 있는 상황에서  
- 원핫 인코딩 : 8개의 컬럼 생성  
- 이진 인코딩 : $\log_28=3$ 개의 컬럼 생성

이 또한 소관경찰서와 소관지역이 이미 number로 입력되어 있기에 실제 object형인 **범죄발생지**에 대해서만 진행하면 된다. 

```python
# 이진 인코딩은 바로 바꾸면 오류가 남.
binary_encoder = BinaryEncoder()
base_train['범죄발생지'] = binary_encoder.fit_transform(base_train[['범죄발생지']])
validation[['범죄발생지']] = binary_encoder.transform(validation[['범죄발생지']]) # val/test는 학습된 인코더로 변환만 
test[['범죄발생지']] = binary_encoder.transform(test[['범죄발생지']])
base_train[['범죄발생지']]
```

```python
# 이런 식으로 인코딩되는 것이기 때문에 바로 대입이 안되는 것.
binary_encoder.fit_transform(base_train[['범죄발생지']]) # 다중 공선성 경우 생각해야함.
```

```python
base_train = pd.concat([base_train.iloc[:,:-1], binary_encoder.fit_transform(base_train['범죄발생지']), base_train.iloc[:,-1]], axis = 1) # TARGET을 마지막으로 하기 위해
validation = pd.concat([validation.iloc[:,:-1], binary_encoder.transform(validation['범죄발생지']), validation.iloc[:,-1]], axis = 1)
test = pd.concat([test.iloc[:,:-1], binary_encoder.transform(test['범죄발생지']), test.iloc[:,-1]], axis = 1)

base_train.drop(columns=['범죄발생지'], inplace = True)
validation.drop(columns=['범죄발생지'], inplace = True)
test.drop(columns=['범죄발생지'], inplace = True)
base_train
```

```python
# 이상치 처리한 데이터셋도 적용.
# 반드시 base의 transform으로
IQR_delete = pd.concat([IQR_delete.iloc[:,:-1], binary_encoder.transform(IQR_delete['범죄발생지']), IQR_delete.iloc[:,-1]], axis = 1)
IQR_weight = pd.concat([IQR_weight.iloc[:,:-1], binary_encoder.transform(IQR_weight['범죄발생지']), IQR_weight.iloc[:,-1]], axis = 1)
Isolation_Forest_delete = pd.concat([Isolation_Forest_delete.iloc[:,:-1], binary_encoder.transform(Isolation_Forest_delete['범죄발생지']), Isolation_Forest_delete.iloc[:,-1]], axis = 1)
Isolation_Forest_weight = pd.concat([Isolation_Forest_weight.iloc[:,:-1], binary_encoder.transform(Isolation_Forest_weight['범죄발생지']), Isolation_Forest_weight.iloc[:,-1]], axis = 1)

IQR_delete.drop(columns=['범죄발생지'], inplace = True)
IQR_weight.drop(columns=['범죄발생지'], inplace = True)
Isolation_Forest_delete.drop(columns=['범죄발생지'], inplace = True)
Isolation_Forest_weight.drop(columns=['범죄발생지'], inplace = True)
```
